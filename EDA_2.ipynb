Plot(a)- Weekly estimated excess deaths by age group

Which Gestalt principles are in use?

1. Law of Simplicity - There is no unnecessary data presented or highlighted in the plot. Everything is easily readable, uncluttered and properly indented. So it is easy to read the data and get the insights.
2. Law of Proximity - the plot is highlighted with the text like 201X-201Y flu season to give the idea that it is related to that piece of plot.
3. Law of Connectedness/enclosure - A bounding box is created to give us insight that for which time interval they are talking in the plot.
4. Law of Similarity - The color pallete chosen for the plot gives the right preception that similar looking colors have similar properties(or distribution or trend). For eg. Row "All ages in 2019 and 2020 flu season" had similar trend with age group "65 and up".

Advantage of chosen color scheme are:
1. We can easily identify the age group which are highly affected by flu in which season.
2. Which seasons is generally more vulnerable to every age group.
3. It takes care of color blind people as red and green are not used in same plot.

No Trump card

Which Gestalt principles are in use?

1. Law of Proximity: Text of 1980 in 1968-92 plot near to the line allows us to clearly relate the line plot of 1980 with year 1980 among the several other year's plot. Similarily with the text '2020' and '2016' on the 1996-2020 plot.
2. Law  of Figure and Focus: Only the line of 1980 in left plot and only the line of 2020 and 2016 are highlighted in contrast color so it is separating these line from their background lines of other years which is making it easier to focus on that particular one.
3. Law of Focal Point: Only the line of 2020 is marked in RED so it stands out visually in the entire plot and will get more attention by the audience.

Advantage of chosen color scheme are:
1. The dark blue used for highlighting the plot of 1980 and 2016. It serves the purpose of focusing on these years than the rest.
2. Two tone color scheme of blue has helped to clearly separate certain lines from the rest.
3. The red color serves the author's purpose of seeking the audience attention at first glance on the graph.
4. The author has not used red and green color on the same plot, so color blind might not face no issues in visualizing the story, the author wants to convey.

```python

d['Revenue'] = d['Revenue']/1000
d['Profit'] =d['Profit']/1000
d['Profit'] = d["Profit"].astype(int)
d['Revenue'] = d["Revenue"].astype(int)
```

```python
import matplotlib.patches as patches
# basic horizontal stacked bar chart

fig,ax=plt.subplots(figsize=(8,6))
ax.set_xlim([0, 600])
ax.barh(np.arange(10), d['Revenue'], color='blue', label='Revenue', height=0.68, alpha=0.6)
ax.barh(np.arange(10), d['Profit'], left = d['Revenue'], color='blue', label='Profit', height=0.68,alpha=0.4)

#hide x ticks and y ticks
ax.tick_params(axis='x', bottom=False, labelbottom=False)
ws=0
#add count values to bars
for i, rect in enumerate(ax.patches):
    h = rect.get_height() #height of each bar
    w = rect.get_width()#width of each bar
    ws=ws+w
    x = rect.get_x()+ w/1.2 #x coordinate of the text
    y = rect.get_y()+h/2.7 #y coordinate of the text
    label = f"{w}"
    if i in np.arange(0,10): #highlighed bars
        ax.annotate(label, (x, y),color='white',size=12,ha='left')
    else: 
        ax.annotate(label, (x+5, y),color='black',size=12,ha='left')
        ax.annotate(d['percent'][i-10], (600, y), color='black', annotation_clip=False, weight='bold', size=12 )
            
# add total text
ax.annotate('Profit(%)', (600, 10), size=12, weight='bold',annotation_clip=False)

#add y ticks labels
companies= d['Company'] 
ax.set_yticks(np.arange(0,10))
col=['black', 'black', 'black', 'black', 'red','black', 'black', 'black', 'black','black']
ax.tick_params(axis='y',direction='out', pad=175, left=False)
ax.set_yticklabels(companies, ha = 'left', fontsize=12, weight ='bold')

for ytick, color in zip(ax.get_yticklabels(), col):
    ytick.set_color(color)

#add legend
ax.legend(bbox_to_anchor=(0.5, 1), ncol=2, prop={'size': 12}) 
#set background color
ax.set_facecolor("#faf4ed")

#hide border
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['left'].set_visible(False)
ax.spines['bottom'].set_visible(False)

ax.annotate("Revenues (in Billions) Vs Profit (in Billions)", (-250, 12), size=16, color='black', weight='bold',annotation_clip=False)
ax.annotate("Top 10 Companies", (-240, 11.3), size=14, color='grey',annotation_clip=False)

ax.add_patch(
    patches.Rectangle(
        xy=(-250, 3.5),  # point of origin.
        width= 900,
        height= 1,
        linewidth=2,
        edgecolor='red',
        facecolor='#eda882',
        alpha=0.3,
        clip_on=False
    )
)
ax.annotate("Market Leader in Profit(%)", (660, 3.8), size=14, color='red',annotation_clip=False)
```

```python
states = pd.read_csv('states.csv')
states
```

```python
states['Population'].mean()
```

```python
population_belowmean = states[states['Population'] < 6496932]
population_abovemean = states[states['Population'] >= 6496932]
population_abovemean.count(), population_belowmean.count()
```

```python
av_unemp_rate = states['Unemployment_rate'].mean()
av_hom_rate = states['Homicide_rate'].mean()
av_unemp_rate, av_hom_rate
```

```python
highlight_high_unemp_hom = states[(states['Unemployment_rate'] >= 5.152) & (states['Homicide_rate'] >= 5.04)]
highlight_high_unemp_hom
```

```python
sizes_above = population_abovemean['Population']**0.4
sizes_below = population_belowmean['Population']**0.3
sizes_highlight = highlight_high_unemp_hom['Population']**0.4


fig, ax = plt.subplots(figsize=(12,7))
#set s to sizes
ax.scatter(population_abovemean['Homicide_rate'], population_abovemean['Unemployment_rate'], color='#0A5252', s=sizes_above, alpha=.3, label='17 States above Mean Population')
ax.scatter(population_belowmean['Homicide_rate'], population_belowmean['Unemployment_rate'], color='#50C7C7',s=sizes_below, alpha=.3, label='33 States above Mean Population')
ax.scatter(highlight_high_unemp_hom['Homicide_rate'], highlight_high_unemp_hom['Unemployment_rate'], color='red',s=sizes_highlight, alpha=0.7, label='11 States: Highest Concern')
ax.set_xlabel("Homicide Rate (%)", weight='bold',size=12,loc='right')
ax.set_ylabel("UnEmployment Rate (%)", weight='bold',size=12,loc='top')

ax.axhline(av_unemp_rate, c='#0A5252', lw=2.5,alpha=0.9, linestyle='--')
ax.axvline(av_hom_rate, c='red', lw=2.5,alpha=0.3, linestyle='--')

#Louisiana - Highest Homicide Rate -12.4 %
#Connecticut- Highest UnEmployment Rate -7.9 %
ax.annotate("Louisiana - Highest Homicide Rate -12.4%", (13.1, 6.8), size=12, color='red', weight='bold',annotation_clip=False,alpha=0.7)
ax.annotate("Nevada- Top Priority State for Action", (7, 8.2), size=15, color='red', weight='bold',annotation_clip=False,alpha=0.9)
ax.annotate("Av Unemployment Rate: 5.1%", (13.1, 5.1), size=12, color='#0A5252', weight='bold',annotation_clip=False)
ax.annotate("Av Homicide Rate: 5%", (5.2, 2.5), size=12, color='red', weight='bold',annotation_clip=False,alpha=0.5)

ax.legend(bbox_to_anchor=(1.3, -0.25), ncol=3, prop={'size': 11,'weight':'bold'}, frameon=False,loc='lower right') 
#hide the frames, only show x and y axis

ax.annotate("Unemployment Rate and Homicide Rate Statistics", (0, 9.45), size=17, color='black', weight='bold',annotation_clip=False)
ax.annotate("Selection of Top Priority States for Action", (0.3, 9.1), size=13, color='grey',annotation_clip=False)

ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

ax.set_facecolor("#faf4ed")

plt.show()
```
